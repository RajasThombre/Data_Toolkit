**1. What is NumPy, and why is it widely used in Python?**

  ->NumPy is a core Python library for numerical computing. It provides fast, memory-efficient multidimensional arrays and mathematical functions, making it widely used for data analysis, scientific computing, and machine learning.

**2. How does broadcasting work in NumPy?**

  -> Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding smaller arrays to match larger ones without copying data.

**3. What is a Pandas DataFrame?**

  -> A Pandas DataFrame is a two-dimensional, labeled data structure similar to a table in a database or an Excel spreadsheet, capable of holding different data types in columns.

**4. Explain the use of the groupby() method in Pandas**

  -> groupby() is used to split data into groups based on one or more columns, apply aggregation functions (like sum, mean), and combine the results for analysis.

**5. Why is Seaborn preferred for statistical visualizations?**

  -> Seaborn provides high-level, visually appealing statistical plots with minimal code and integrates well with Pandas DataFrames.

**6. What are the differences between NumPy arrays and Python lists?**

  -> NumPy arrays are faster, use less memory, support vectorized operations, and store elements of the same data type, while Python lists are slower and more flexible but less efficient.

**7. What is a heatmap, and when should it be used?**

  -> A heatmap is a graphical representation of data where values are shown using colors. It is commonly used to visualize correlations, matrices, or intensity patterns.

**8. What does the term “vectorized operation” mean in NumPy?**

  -> A vectorized operation performs calculations on entire arrays at once instead of using loops, resulting in faster execution.

**9. How does Matplotlib differ from Plotly?**

  -> Matplotlib is static and best for basic plots, while Plotly is interactive, web-friendly, and supports hover, zoom, and animations.

**10. What is the significance of hierarchical indexing in Pandas?**

  -> Hierarchical indexing (MultiIndex) allows multiple index levels, enabling more complex data representation and easier analysis of higher-dimensional data.

**11. What is the role of Seaborn’s pairplot() function?**

  -> pairplot() creates pairwise scatter plots and distributions to quickly explore relationships between multiple numerical variables.

**12. What is the purpose of the describe() function in Pandas?**

  -> describe() generates summary statistics such as count, mean, standard deviation, min, max, and quartiles for numerical data.

**13. Why is handling missing data important in Pandas?**

 -> Missing data can lead to incorrect analysis and model bias. Pandas provides tools to detect, fill, or remove missing values to maintain data quality.

**14. What are the benefits of using Plotly for data visualization?**

  -> Plotly offers interactive charts, easy dashboard integration, rich visuals, and support for web-based applications.

**15. How does NumPy handle multidimensional arrays?**

  -> NumPy uses the ndarray object, which supports arrays of any dimension and allows efficient slicing, reshaping, and mathematical operations.

**16. What is the role of Bokeh in data visualization?**

  -> Bokeh is used to create interactive, browser-based visualizations suitable for dashboards and real-time data applications.

**17. Explain the difference between apply() and map() in Pandas**

  -> map() works on Series elements individually, while apply() can operate on entire rows or columns of a DataFrame.

**18. What are some advanced features of NumPy?**

  -> Advanced features include broadcasting, memory mapping, linear algebra operations, FFT, random number generation, and masked arrays.

**19. How does Pandas simplify time series analysis?**

  -> Pandas provides built-in datetime indexing, resampling, rolling windows, and time-based slicing, making time series analysis easier.

**20. What is the role of a pivot table in Pandas?**

  -> A pivot table summarizes data by aggregating values based on rows, columns, and functions like sum or mean.

**21. Why is NumPy’s array slicing faster than Python’s list slicing?**

  -> NumPy slicing creates views instead of copies and operates in compiled C code, making it much faster.

**22. What are some common use cases for Seaborn?**

 -> Common use cases include data exploration, statistical analysis, correlation analysis, regression plots, and distribution visualization.

```python
# Q1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# Q2. Write a Pandas script to find the mean of a specific column in a DataFrame

import pandas as pd

df = pd.DataFrame({
    "marks": [70, 80, 90, 100]
})

mean_value = df["marks"].mean()
print(mean_value)
```

```python
# Q3. Create a scatter plot using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.show()
```

```python
# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [2, 4, 6, 8],
    "C": [5, 6, 7, 8]
})

corr_matrix = data.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Q5. Generate a bar plot using Plotly

import plotly.express as px

data = {
    "Category": ["A", "B", "C"],
    "Values": [10, 20, 15]
}

fig = px.bar(data, x="Category", y="Values", title="Bar Plot")
fig.show()
```

```python
# Q6. Create a DataFrame and add a new column based on an existing column

import pandas as pd

df = pd.DataFrame({
    "salary": [30000, 40000, 50000]
})

df["bonus"] = df["salary"] * 0.10
print(df)
```

```python
# Q7. Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
# Q8. Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

```python
# Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

df = pd.DataFrame({
    "age": [18, 25, 30, 15]
})

filtered_df = df[df["age"] > 20]
print(filtered_df)
```

```python
# Q10. Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40]

sns.histplot(data, bins=5)
plt.title("Histogram")
plt.show()
```

```python
# Q11. Perform matrix multiplication using NumPy

import numpy as np

a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

result = np.dot(a, b)
print(result)
```

```python
# Q13. Create a 3D scatter plot using Plotly

import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 3, 4],
    "y": [10, 20, 30, 40],
    "z": [5, 15, 25, 35]
})

fig = px.scatter_3d(df, x="x", y="y", z="z", title="3D Scatter Plot")
fig.show()
```
